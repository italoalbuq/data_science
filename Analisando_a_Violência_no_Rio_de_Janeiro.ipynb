# Analisando a Violência no Rio de Janeiro

A violência no Rio de Janeiro é uma das maiores do País, e sempre ganhou destaque nas mídias brasileiras.

O governo do RJ, disponibiliza dados pelo [portal ISPDados](http://www.ispdados.rj.gov.br/). Farei uma análise exploratória com indicadores estatístico da violência do Rio de Janeiro.

## Obtenção dos Dados


Os dados foram obtidos pelo [portal ISPDados](http://www.ispdados.rj.gov.br/),que é uma página de Dados Abertos do Instituto de Segurança Pública. Nesta página, você poderá acessar as bases de dados de registros criminais e de atividade policial do estado do Rio de Janeiro.

O presente processo de revisão de dados é mais uma etapa do compromisso do ISP com a transparência e a qualidade dos dados oficiais da Segurança Pública do Estado do Rio de Janeiro.

O *dataset* utilizado aqui, pode ser baixado diretamente [neste link](https://raw.githubusercontent.com/carlosfab/curso_data_science_na_pratica/master/modulo_02/violencia_rio.csv).

## Informações Estatísticas da Violência no Rio de Janeiro

* Resumo estatístico do DataFrame.

```
# resumo estatístico do DataFrame
df.describe()
```

* Médias das seguintes variáveis (colunas):
  * `roubo_veiculo`
  * `furto_veiculos`
  * `recuperacao_veiculos`

```
# médias de algumas colunas
print('Média de veículos roubados: ', df.roubo_veiculo.mean())
print('Média de veículos furtados: ', df.furto_veiculos.mean())
print('Média de veículos recuperados: ', df.recuperacao_veiculos.mean())
```

* Porcentagem de carros recuperados em relação aos carros roubados + carros furtados:
  * $\frac{\text{recuperacao_veiculos}}{\text{roubo_veiculo} + \text{furto_veiculos}}$

```
# porcentagem de carros recuperados em relação aos carros roubados + carros furtados
df.recuperacao_veiculos.mean() / (df.furto_veiculos.mean() + df.roubo_veiculo.mean())
```

* Valores máximos (`max`) e mínimos (`min`) da coluna `hom_doloso`

```
# valores max e min de hom_doloso
print('Valor máximo de Homicídio Doloso: ', df.hom_doloso.max())
print('Valor mínimo de Homícidio Doloso: ', df.hom_doloso.min())
```
